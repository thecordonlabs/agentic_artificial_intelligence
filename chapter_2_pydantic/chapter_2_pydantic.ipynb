Pydantic 

```python
from pydantic import BaseModel
class User(BaseModel):
    first_name:str
    middle_name:str
    last_name:str
    age:int
    email:str
user_1 = User(
    first_name="jash",
    middle_name="manish",
    last_name='upadhyay',
    age=21,
    email="the.cordon.labs@gmail.com"
)
print(user_1)
print(type(user_1))
# user_2 = User(
#     first_name="jash1",
#     middle_name="manish1",
#     last_name='upadhyay1',
#     age="",
#     email="the.cordon.labs@gmail.com"
# )
# above commented will throw validation error
```

```python
# another way of validation
from dataclasses import dataclass
@dataclass
class User1():
    first_name:str
    middle_name:str
    last_name:str
    age:int
    email:str   
user_1 = User1(
    first_name="jash",
    middle_name="manish",
    last_name='upadhyay',
    age=21,
    email="the.cordon.labs@gmail.com"
)
print(user_1)
print(type(user_1))
# user_2 = User1(
#     first_name="jash1",
#     middle_name="manish1",
#     last_name='upadhyay1',
#     age="",
#     email="the.cordon.labs@gmail.com"
# )
# above commented will throw validation error 
```
