### Pydantic 

```python
from pydantic import BaseModel
class User(BaseModel):
    first_name:str
    middle_name:str
    last_name:str
    age:int
    email:str
user_1 = User(
    first_name="jash",
    middle_name="manish",
    last_name='upadhyay',
    age=21,
    email="the.cordon.labs@gmail.com"
)
print(user_1)
print(type(user_1))
# user_2 = User(
#     first_name="jash1",
#     middle_name="manish1",
#     last_name='upadhyay1',
#     age="",
#     email="the.cordon.labs@gmail.com"
# )
# above commented will throw validation error
```

```python
# another way of validation
from dataclasses import dataclass
@dataclass
class User1():
    first_name:str
    middle_name:str
    last_name:str
    age:int
    email:str   
user_1 = User1(
    first_name="jash",
    middle_name="manish",
    last_name='upadhyay',
    age=21,
    email="the.cordon.labs@gmail.com"
)
print(user_1)
print(type(user_1))
# user_2 = User1(
#     first_name="jash1",
#     middle_name="manish1",
#     last_name='upadhyay1',
#     age="",
#     email="the.cordon.labs@gmail.com"
# )
# above commented will throw validation error 
```

```python
# optional fields(eg. age:Optional[type]): type for the field can be the "type" or None
from typing import Optional
class Customer(BaseModel):
    full_name:str
    age:Optional[int]
cust_1 = Customer(full_name="jash_upadhyay",age=None)
print(cust_1)
cust_2 = Customer(full_name="nil_zero",age=21)
print(cust_2)
# cust_3 = Customer(full_name="nil_zero",age="nil_zero")
# print(cust_3)
# above commented code will throw validation error
```

```python
# nested models
class Buyer(BaseModel):
    full_name:str
    age:int
class Product(BaseModel):
    product_name:str
    buyer:Buyer
product_1 = Product(
    product_name="mixer",
    buyer={
        "full_name":"jash upadhyay",
        "age":21
    }
)
```

```python
# pydantic fields: allows to add more validation rules
from pydantic import Field
class Employee(BaseModel):
    full_name:str=Field(max_length=30) # string can only be 30 characters long
    age:int=Field(gt=17,le=50) # greater than 17, less than equal to 50
emp_1 = Employee(full_name="abcde",age=21)
# emp_2 = Employee(full_name="aaaaaa aaaaaa aaaaaa aaaaaa aa",age=17)
# above commented will throw validation error
```
